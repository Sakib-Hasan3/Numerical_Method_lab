```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**3 - 4*x - 9 
```

```python
def bisection(a, b, n=10):
    print("Iter\t   a\t\t   b\t\t   c\t\t f(c)")
    print("-------------------------------------------------------------")
    for i in range(1, n+1):
        c = (a + b) / 2
        print(f"{i}\t{a:.5f}\t{b:.5f}\t{c:.5f}\t{f(c):.5f}")
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c
```

```python
a, b = 2, 3
root = bisection(a, b)
print("Root ≈", root)
```

```python
x = np.linspace(1, 4, 400)
y = f(x)
plt.plot(x, y, label="f(x)")
plt.axhline(0, color="black")
plt.axvline(root, color="red", linestyle="--", label=f"Root ≈ {root:.4f}")
plt.scatter(root, f(root), color="red")
plt.legend()
plt.title("Bisection Method (Easy Code)")
plt.show()
```
